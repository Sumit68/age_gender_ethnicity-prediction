```python

A = os.listdir('/kaggle/input/utkface-new/UTKFace')
```

```python
Path = '/kaggle/input/utkface-new/UTKFace'
```

```python
labels = np.array([ float(i.split('_')[0]) for i in A])
```

```python
gc.collect()
```

```python
import cv2
```

```python
#../input/utkface-new/UTKFace/100_0_0_20170112213500903.jpg.chip.jpg
```

```python
def format_image(i):
  img = cv2.imread(Path + '/' + i)
  img = cv2.resize(img,(128,128))
  return img
```

```python
data = np.array([format_image(i) for i in A])
```

```python
data.shape
```

```python
gc.collect()
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(data,labels,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.applications import MobileNet
import  tensorflow.keras.layers as L
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
                  MobileNet(input_shape=(128,128,3),
                            include_top=False,
                            pooling='avg',
                            weights='imagenet'),
                             L.Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mae',metrics='mae')
```

```python
test_y
```

```python
model.fit(
    train_x,
    train_y,
    batch_size=64,
    epochs=10,
    validation_data=(test_x,test_y)
)
```
