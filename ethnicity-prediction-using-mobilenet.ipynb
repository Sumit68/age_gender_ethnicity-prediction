```python
A = os.listdir('/kaggle/input/utkface-new/UTKFace')
```

```python
Path = '/kaggle/input/utkface-new/UTKFace'
```

```python
# Errors in some Filename
labels = []
for i in A:
    if i.split('_')[2] == '20170109142408075.jpg.chip.jpg':
        labels.append(1)
    elif i.split('_')[2] == '20170109150557335.jpg.chip.jpg':
        labels.append(3)
    elif i.split('_')[2] == '20170116174525125.jpg.chip.jpg':
        labels.append(1)
    else:
        labels.append(int(i.split('_')[2]))
        
    
 
    
    
```

```python
labels
```

```python
gc.collect()
```

```python
import cv2
```

```python
def format_image(i):
  img = cv2.imread(Path + '/' + i)
  img = cv2.resize(img,(128,128))
  return img
```

```python
data = np.array([format_image(i) for i in A])
```

```python
data.shape
```

```python
labels = np.array(labels)
```

```python
labels.shape
```

```python
gc.collect()
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(data,labels,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.applications import MobileNet
import  tensorflow.keras.layers as L
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
                  MobileNet(input_shape=(128,128,3),
                            include_top=False,
                            pooling='avg',
                            weights='imagenet'),
                            L.Dense(5, activation="softmax")
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

model.fit(
    train_x,
    train_y,
    batch_size=64,
    epochs=9,
    validation_data=(test_x,test_y)
)
```
